### 2. How does broadcasting work in NumPy?

**Broadcasting** in NumPy allows **arrays of different shapes** to be used in **arithmetic operations** without explicit copying or looping.

####  **The Basic Idea:**

Broadcasting automatically **expands the smaller array** to match the shape of the larger one **without actually copying data**, making operations **memory-efficient and fast**.

#### ðŸ” **Rules of Broadcasting:**

When operating on two arrays, NumPy compares their shapes **element-wise** from **right to left**. Two dimensions are **compatible** when:

1. They are **equal**, or
2. One of them is **1**

If the dimensions are incompatible, NumPy will raise an error.

#### âœ… **Examples:**

##### 1. **Adding Scalar to Array:**

```python
import numpy as np

a = np.array([1, 2, 3])
b = 10

print(a + b)  # [11 12 13]
```

* Scalar `b` is "stretched" to `[10, 10, 10]`.

##### 2. **Adding 2D and 1D Arrays:**

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)
```

**Output:**

```
[[11 22 33]
 [14 25 36]]
```

* `b` is broadcast to match the shape of `a` (2x3).

##### 3. **Incompatible Shapes Example:**

```python
a = np.array([1, 2, 3])
b = np.array([[1], [2]])

# print(a + b)  # âŒ This would raise a broadcasting error.
```

Letâ€™s say you have:

```
Array A: (3, 1)
Array B: (1, 4)

Result after broadcasting â†’ shape: (3, 4)
```

Each array is stretched without copying data, and element-wise operations can be applied.

### 6. What are the differences between NumPy arrays and Python lists?

While both **NumPy arrays** and **Python lists** can store collections of data, they are **fundamentally different** in how they work, especially for **numerical and scientific computing**.

#### **Comparison Table:**

| Feature                     | **NumPy Array (`np.array`)**                       | **Python List (`list`)**                   |
| --------------------------- | -------------------------------------------------- | ------------------------------------------ |
| **Data Type**               | Homogeneous (all elements of the same type)        | Heterogeneous (can store mixed types)      |
| **Speed**                   | Much faster (optimized C backend)                  | Slower (interpreted Python)                |
| **Memory Efficiency**       | Uses less memory                                   | Consumes more memory                       |
| **Element-wise Operations** | Supports vectorized operations                     | Requires loops for element-wise operations |
| **Functionality**           | Rich mathematical and linear algebra operations    | Limited built-in numerical capabilities    |
| **Slicing & Broadcasting**  | Advanced slicing, broadcasting, masking            | Basic slicing only                         |
| **Multidimensional**        | Supports multi-dimensional arrays (matrix, tensor) | Needs nested lists                         |

#### **Example:**

##### 1. **Python List (No vectorized operations)**

```python
a = [1, 2, 3]
b = [4, 5, 6]
result = [i + j for i, j in zip(a, b)]
print(result)  # [5, 7, 9]
```

##### 2. **NumPy Array (Clean and fast)**

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)  # [5 7 9]
```

#### **When to Use What?**

| Use Case                         | Recommended |
| -------------------------------- | ----------- |
| Basic storage / general use      | Python List |
| Numerical / scientific computing | NumPy Array |

### 8. What does the term "vectorized operation" mean in NumPy?


A **vectorized operation** in NumPy means applying an **operation to entire arrays** (vectors, matrices, etc.) **without writing explicit loops**.

#### **Key Idea:**

Instead of processing elements **one-by-one** in a loop (like in pure Python), NumPy uses **low-level optimized C code** to apply operations **to whole arrays at once**.

#### **Example**

##### Without Vectorization (Python loop):

```python
a = [1, 2, 3, 4]
b = [2*x for x in a]
print(b)  # [2, 4, 6, 8]
```

##### With Vectorization (NumPy):

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = 2 * a
print(b)  # [2 4 6 8]
```

> No loop, faster execution, cleaner code!

#### **Benefits of Vectorized Operations:**

| Benefit                 | Explanation                                      |
| ----------------------- | ------------------------------------------------ |
| âš¡ **Faster**            | Uses compiled C under the hood â€” no Python loops |
| ðŸ§¼ **Cleaner code**     | No need for `for` loops or list comprehensions   |
| ðŸ’¾ **Memory-efficient** | Reduces overhead from Python objects             |
| ðŸ“ˆ **Scalable**         | Easily handles large datasets                    |


#### **Examples of Vectorized Operations:**

| Operation      | Example (`a = np.array([1, 2, 3])`) | Result              |
| -------------- | ----------------------------------- | ------------------- |
| Addition       | `a + 10`                            | `[11 12 13]`        |
| Multiplication | `a * 2`                             | `[2 4 6]`           |
| Power          | `a ** 2`                            | `[1 4 9]`           |
| Comparison     | `a > 1`                             | `[False True True]` |

### 10. What is the significance of hierarchical indexing in Pandas?


**Hierarchical indexing** (also called **MultiIndexing**) in Pandas allows you to have **multiple levels of indices** on rows or columns. This gives you a **more powerful and flexible way** to organize, analyze, and manipulate high-dimensional data in a 2D structure like a DataFrame.

#### **Why Use Hierarchical Indexing?**

| Purpose                        | Benefit                                                       |
| ------------------------------ | ------------------------------------------------------------- |
| ðŸ“š Organize complex data       | Efficiently represent 2D or 3D data in tabular form           |
| ðŸ” Easy subsetting & filtering | Allows filtering and slicing based on multiple levels of keys |
| ðŸ“Š Group-based analysis        | Works well with `groupby()` and pivot tables                  |
| ðŸ”„ Reshaping data              | Simplifies operations like stacking and unstacking            |

#### **Example:**

```python
import pandas as pd

# Create a MultiIndex DataFrame
data = {
    'Sales': [250, 450, 300, 500],
    'Profit': [20, 60, 30, 80]
}
index = pd.MultiIndex.from_tuples([
    ('2024', 'Q1'),
    ('2024', 'Q2'),
    ('2025', 'Q1'),
    ('2025', 'Q2')
], names=['Year', 'Quarter'])

df = pd.DataFrame(data, index=index)
print(df)
```

**Output:**

```python
              Sales  Profit
Year Quarter                
2024 Q1         250      20
     Q2         450      60
2025 Q1         300      30
     Q2         500      80
```

#### **Operations Made Easy with Hierarchical Indexing:**

##### Access a subgroup:

```python
df.loc['2024']
```

##### Slice by multiple levels:

```python
df.loc[('2025', 'Q1')]
```

##### Swap or sort levels:

```python
df.swaplevel()
df.sort_index(level='Quarter')
```

### 13. Why is handling missing data important in Pandas?


Handling **missing data** is crucial because it can lead to **inaccurate analyses**, **misleading visualizations**, and **broken machine learning models**.

#### **Key Reasons Why It's Important:**

| Reason                               | Explanation                                                                |
| ------------------------------------ | -------------------------------------------------------------------------- |
| **Prevents Data Bias**            | Missing values can skew your results or statistical summaries              |
| **Avoids Errors in Calculations** | Operations like mean, sum, or ML models may break or give NaN as output    |
| **Improves Visualization**        | Plots may fail or misrepresent data if missing values are not handled      |
| **Essential for Modeling**        | Most ML algorithms (like scikit-learn models) do **not accept NaN** values |
| **Data Integrity**                | Ensures consistency and completeness in reporting or decision making       |

#### **Common Methods to Handle Missing Data in Pandas:**

##### 1. **Detect Missing Values**

```python
df.isnull()       # Returns True/False for each cell
df.isnull().sum() # Count of missing values per column
```

##### 2. **Remove Missing Data**

```python
df.dropna()             # Drop rows with any NaN values
df.dropna(axis=1)       # Drop columns with any NaN values
```

##### 3. **Fill Missing Data**

```python
df.fillna(0)            # Fill with a constant
df.fillna(method='ffill')  # Forward fill
df.fillna(method='bfill')  # Backward fill
df['col'].fillna(df['col'].mean())  # Fill with column mean
```

#### **Example:**

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, None, 92, None]
}

df = pd.DataFrame(data)

# Fill missing scores with the mean
df['Score'] = df['Score'].fillna(df['Score'].mean())
```

### 15. How does NumPy handle multidimensional arrays?


NumPy uses a powerful data structure called the **ndarray** (N-dimensional array) to handle **multidimensional data** â€” ranging from simple vectors (1D) to matrices (2D) and even higher dimensions like 3D or more.

#### **What Is an ndarray?**

* A NumPy array object that can have **any number of dimensions**
* All elements in an array are of the **same data type**
* Supports fast, element-wise operations across dimensions

#### **Creating Multidimensional Arrays**

```python
import numpy as np

# 1D array (vector)
a = np.array([1, 2, 3])

# 2D array (matrix)
b = np.array([[1, 2, 3], [4, 5, 6]])

# 3D array (tensor)
c = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```

#### **Important Attributes:**

| Attribute | Description               | Example (for `b`)        |
| --------- | ------------------------- | ------------------------ |
| `.ndim`   | Number of dimensions      | `b.ndim â†’ 2`             |
| `.shape`  | Tuple of dimensions       | `b.shape â†’ (2, 3)`       |
| `.size`   | Total number of elements  | `b.size â†’ 6`             |
| `.dtype`  | Data type of the elements | `b.dtype â†’ int64` (e.g.) |

#### **Accessing Elements in Multidimensional Arrays:**

```python
print(b[0, 1])   # Access 1st row, 2nd column â†’ 2
print(c[1, 0, 1])  # Access from 3D â†’ 6
```

#### **Operations on Multidimensional Arrays:**

* Element-wise operations work seamlessly:

```python
b + 10       # Adds 10 to each element
b.T          # Transpose for 2D arrays
np.sum(b, axis=0)  # Sum along columns
np.sum(b, axis=1)  # Sum along rows
```

### 17. Explain the difference between apply() and map() in Pandas.


#### **1. `map()` â€“ Element-wise for Series Only**

* Best for **simple element-wise transformations** on a single column.

##### Example:

- On a Series:

    ```python
    import pandas as pd

    s = pd.Series([1, 2, 3, 4])

    # Multiply each value by 10
    s.map(lambda x: x * 10)
    ```

- On a dictionary (useful for value mapping):

    ```python
    s.map({1: 'One', 2: 'Two'})
    ```

#### **2. `apply()` â€“ More Flexible, Works on Series & DataFrames**

* Can be used on **DataFrames** (row/column level) or **Series**
* Handles **custom logic** more easily

- On a Series:

    ```python
    s.apply(lambda x: x ** 2)
    ```

- On a DataFrame:

    ```python
    df = pd.DataFrame({
        'A': [1, 2, 3],
        'B': [10, 20, 30]
    })

    # Apply sum across rows
    df.apply(lambda row: row['A'] + row['B'], axis=1)

    # Apply sum across columns
    df.apply(sum, axis=0)
    ```

#### ðŸ“Œ **Summary Table:**

| Feature        | `map()`                                | `apply()`                                |
| -------------- | -------------------------------------- | ---------------------------------------- |
| Data Type      | Pandas Series only                     | Series & DataFrame                       |
| Operation Type | Element-wise                           | Element-wise, row-wise, column-wise      |
| Accepts        | Function, dictionary, Series           | Any Python function                      |
| Returns        | Series                                 | Series, DataFrame, or scalar             |
| Use Case       | Value mapping or simple transformation | Complex functions, multiple column logic |

### 18. What are some advanced features of NumPy?


#### **1. Broadcasting**

Allows operations on arrays of **different shapes** without explicit replication.

```python
import numpy as np

a = np.array([[1], [2], [3]])   # shape (3,1)
b = np.array([10, 20, 30])      # shape (3,)

print(a + b)  # Automatically broadcasts to shape (3,3)
```

#### **2. Vectorization**

Replaces Python loops with **fast, element-wise operations**, leveraging low-level C for performance.

```python
x = np.arange(1_000_000)
y = x ** 2  # No explicit loop, very fast
```

#### **3. Advanced Indexing & Boolean Masking**

Extract elements based on conditions or index arrays.

```python
arr = np.array([10, 20, 30, 40])
print(arr[[0, 2]])         # index array
print(arr[arr > 25])       # boolean mask
```

#### **4. Memory Views with `ndarray.view()`**

Create a **view** (not a copy) of an array, saving memory.

```python
a = np.array([1, 2, 3])
b = a.view()
b[0] = 100
print(a)  # a is also changed
```

#### **5. Linear Algebra Operations**

NumPy has a full suite of **matrix operations**:

```python
from numpy.linalg import inv, eig, det

A = np.array([[1, 2], [3, 4]])
inv_A = inv(A)
eigen_vals, eigen_vecs = eig(A)
```

#### **6. Broadcasting Tricks with `np.newaxis` or `reshape()`**

Add dimensions to arrays to make broadcasting easier:

```python
a = np.array([1, 2, 3])         # shape (3,)
b = a[:, np.newaxis]            # shape (3,1)
```

#### **7. Structured Arrays (Custom dtypes)**

Allows heterogeneous data in a single ndarray.

```python
dt = np.dtype([('name', 'U10'), ('age', 'i4')])
people = np.array([('Alice', 25), ('Bob', 30)], dtype=dt)
print(people['name'])
```

#### **8. Memory Mapping Large Files with `memmap`**

Read large arrays from disk without loading everything into memory:

```python
data = np.memmap('large_array.dat', dtype='float32', mode='r', shape=(1000, 1000))
```

#### **9. Performance Tuning with `np.vectorize` and `ufuncs`**

Create vectorized wrappers for Python functions:

```python
def square(x): return x ** 2
vfunc = np.vectorize(square)
vfunc([1, 2, 3])
```

#### **10. Random Number Generation with `np.random`**

Supports complex distributions and reproducibility:

```python
np.random.seed(42)
samples = np.random.normal(loc=0, scale=1, size=1000)
```

Aggregate time series to different frequencies (e.g., daily to monthly).

```python

df.resample('15T').interpolate()
```

#### **4. Shifting and Lagging**

Use `.shift()` to create lag features or forecast targets.

```python
df['sales_lag1'] = df['sales'].shift(1)
```

Calculate rolling statistics like moving average or rolling sum.

df['rolling_mean'] = df['sales'].rolling(window=7).mean()

Pandas can easily detect and fill missing time periods:

df.index = df.index.tz_localize('UTC').tz_convert('Asia/Kolkata')

### 20. What is the role of a pivot table in Pandas?


A **pivot table** in Pandas is a powerful tool used to **summarize, group, and reorganize** data in a DataFrame. It is especially useful for **data analysis**, **reporting**, and exploring patterns across categories.

#### **Key Purpose of a Pivot Table:**

To **aggregate data** (e.g., sum, mean, count) based on **one or more keys**, just like pivot tables in Excel.

**Syntax:**

```python
df.pivot_table(values='column_to_aggregate',
               index='row_grouping_column',
               columns='column_grouping_column',
               aggfunc='function',  # e.g., sum, mean, count
               fill_value=0)
```

**Example:**

```python
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Revenue': [2500, 2700, 1000, 1100, 2000, 2100]
}

df = pd.DataFrame(data)

pivot = df.pivot_table(values='Revenue', index='Department', columns='Month', aggfunc='sum', fill_value=0)
print(pivot)
```

**Output:**

```
Month       Feb   Jan
Department           
HR          1100  1000
IT          2100  2000
Sales       2700  2500
```

#### ðŸ’¡ **Common Use Cases:**

| Use Case                              | Description                                   |
| ------------------------------------- | --------------------------------------------- |
| Group data                            | Summarize by category (e.g., sales by region) |
| Cross-tabulation                      | Show values based on multiple dimensions      |
| Aggregation                           | Mean, sum, count of values                    |
| Missing data handling                 | Fill with 0 or any default using `fill_value` |
| Data transformation for visualization | Feed into heatmaps, bar charts, etc.          |

#### ðŸ” **Difference from `groupby()`**

| Feature         | `pivot_table()`                 | `groupby()`                  |
| --------------- | ------------------------------- | ---------------------------- |

| Multiple Axes   | Supports both index and columns | Only groups by index         |
| Excel-like View | Yes                             | No                           |

### 21. Why is NumPy's array slicing faster than Python's list slicing?


NumPyâ€™s array slicing is significantly faster than Pythonâ€™s list slicing because **NumPy arrays are implemented in C and optimized for numerical computation**. Here's a breakdown of **why** it's faster:

#### âš™ï¸ **1. NumPy Uses Contiguous Memory Blocks**

* NumPy arrays are stored in **contiguous memory**, meaning elements are laid out side-by-side in memory.
* Python lists, on the other hand, are arrays of pointers to objects scattered in memory.

âœ… **Result:** NumPy can access and slice memory using low-level pointer arithmetic, making it **much faster**.

#### ðŸ§  **2. NumPy Slicing Returns Views, Not Copies**

* When you slice a NumPy array, it usually returns a **view** of the original data â€” no copying involved.
* Python list slicing always creates a **new list** with copied elements.

âœ… **Result:** NumPy avoids overhead and memory duplication.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = a[1:3]
b[0] = 100
print(a)  # [  1 100   3   4] â€” reflects the change
```

#### ðŸ› ï¸ **3. NumPy Operations Are Vectorized and Low-Level**

* NumPy's operations (including slicing) are implemented in **compiled C code**.
* Python lists rely on **interpreted, high-level Python loops**.

âœ… **Result:** NumPy's slicing is executed at **machine speed**.

#### ðŸ“Š **4. Fixed Data Types in NumPy Arrays**

* NumPy arrays store elements of **the same data type**, allowing CPU-friendly, SIMD-optimized operations.
* Python lists can hold mixed types, requiring more overhead for type checking.

âœ… **Result:** More **efficient memory layout and faster processing**.

#### â±ï¸ **Benchmark Example:**

```python
import time
import numpy as np

a = list(range(10**6))
b = np.arange(10**6)

# Python list slicing
start = time.time()
x = a[100:100000]
print("List slicing time:", time.time() - start)

# NumPy array slicing
start = time.time()
y = b[100:100000]
print("NumPy slicing time:", time.time() - start)
```

Youâ€™ll typically see that NumPy slicing is **at least several times faster**.
